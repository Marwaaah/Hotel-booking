### Explore missing values

```python
data.isnull().sum()
```

```python
number_missing = data.isnull().sum().sort_values(ascending=False)
pct_column = (data.isnull().sum() / len(data) * 100).round(2).astype('str')+' %'
pct_total = (data.isnull().sum()/data.isna().sum().sum()*100).round(2).astype('str')+' %'
missing_values = pd.concat([number_missing, pct_column, pct_total], axis=1, keys=['Number_Missing_Values', 'PCT_Missing_in_Column','PCT_of_all_Missing'])
   
print('\nMISSING VALUES IN',data.shape[0],'ROWS:')    
all_missing = missing_values.loc[missing_values['Number_Missing_Values'] > 0]
all_missing
```

### Visualisation of missing data

```python
plt.figure(figsize=(10,5))
sns.heatmap(data.isnull(), yticklabels = False, cmap = 'plasma')
plt.title('Null Values in Data Frame')
```
